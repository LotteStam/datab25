# SQL oefeningen

Om de SQL opgaven te kunnen oefenen, werken we aan opdrachten in een Jupyter Notebook. In code-cellen kunnen we instructies typen, bij dit practicum zijn dat vaak SQL queries. De code in een cel kun je uitvoeren door de cel te selecteren en `SHIFT-ENTER` in te toetsen. Als de code uitvoer geeft, wordt die onder de cel getoond.

# Case: Project

Voor deze tutorial maken we gebruik van de tabel medewerker van de database 'PROJECT' die ook in de reader wordt gebruikt. Onderstaande queries zijn alleen ter illustratie hoe de SQL Notebooks werken, in volgende notebooks worden de SQL queries geoefend.

Je kunt alleen werken met een database als je deze connect, in het notebook staat daarom een %connect <database> cel, die moet je uitvoeren.

````python
%connect project
````

#### Tutorial Opdracht 1

**Geef de naam en woonplaats van alle medewerkers.**

De opdracht geeft aan wat er moet worden getoond. Je kunt zien welke gegevens er uit een correcte query komen door de onderstaande cel uit te voeren, waarin %output met het opdrachtnummer staat.

Je kunt een uitwerking voor de opdracht uitproberen. Indien je eerst wilt kijken hoe de tabellen in de database heten, kun je gebruik maken van %tables.

#### SQL query proberen
Om in een notebook iets met SQL uit te proberen of commentaar op te nemen kun je een nieuwe cel tussenvoegen (zie boven). Als in de cel programmacode komt te staan moet je het *cell type* op *code* zetten. Om een SQL query uit te kunnen voeren zet je `%%sql` op de eerste regel met een query daaronder. Zoals bijvoorbeeld in de cell hieronder is gedaan. Maak zelf ook een nieuwe cel en zet daarin een query in die je uitvoert.

````python
%%sql
select * 
from medewerker
````

#### Tijdelijke cel weer verwijderen
Als je een tijdelijke cel niet meer nodig hebt kun je die verwijderen door in CEL-MODUS een 'x' te typen (of in INSERT-MODUS ESC-'x').

#### Opdrachten beantwoorden
Om de opdracht te beantwoorden gebruik je %%uitwerking OPDRACHTNUMMER met daaronder de query. Typ in de cel hieronder het onderstaande in en run de cel met SHIFT-ENTER:
```
%%uitwerking 1
SELECT naam, woonplaats
FROM medewerker
```

````python
%%uitwerking 1
SELECT naam, woonplaats
FROM medewerker
````

#### Automatisch feedback op jouw antwoorden
Jouw antwoorden worden automatisch gecontroleerd. Als je een antwoord goed hebt, dan kleur de titelbalk groen (zie boven).

Let goed op het aantal kolommen dat gevraagd wordt. Als je teveel of te weinig kolommen opgeeft dan zie je geen groene balk maar wel een rode foutmelding onderaan.

````python
%%uitwerking 1
select naam 
from medewerker
limit 2
````

Als het aantal rijen niet overeenkomt met de gewenste uitvoer dan kleurt de titelbalk ook rood. Let vooral op rijen met een NULL waarde (getoond als een leeg vakje), voor SQL is dat ook een rij.

````python
%%uitwerking 1
select naam, woonplaats 
from medewerker 
limit 2
````

De volgorde van de rijen maakt meestal niet uit. Bij de opgaven waarin de rijen gesorteerd moeten worden wordt dat (nog) niet gecontroleerd, dus kijk zelf of de volgorde dan correct is. De volgende query wordt daarom ook goed gekeurd.

````python
%%uitwerking 1
select naam, woonplaats 
from medewerker 
order by woonplaats
````

Als het aantal rijen en kolommen klopt, dan worden de rijen gecontroleerd. Rijen die niet kloppen kleuren dan rood om je te helpen de fout op te sporen.

````python
%%uitwerking 1
select naam, "Leiden" 
from medewerker
````

Tot slot, kun je natuurlijk ook fouten maken in je query. Dan wordt de foutmelding van SQL getoond om je te helpen fouten eruit te halen. In onderstaande query staat bijvoorbeeld een kolom geboorteplaats die niet in de tabel medewerker staat.

````python
%%uitwerking 1
select naam, geboorteplaats 
from medewerker
````

De foutmeldingen zijn niet altijd even duidelijk. In onderstaande voorbeeld is de fout natuurlijk dat er FFFROM staat in plaats van FROM, maar vaak zit de fout wel in de buurt van het woord waar SQL over klaagt.

````python
%%uitwerking 1
select naam, geboorteplaats 
FFFROM medewerker
````
