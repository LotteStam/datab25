## SUBSELECT

We mogen in SQL ook gebruik maken van 'nested queries', dit heet een subselect. Je kunt bijvoorbeeld de resultaten van een subselect gebruiken in de WHERE-clausule van een andere query. Zet daarvoor de SUBSELECT tussen (), en je kunt met de IN- en NOT IN-operatoren bepalen of een waarde voorkomt in de subselect. Maar op een soortgelijke manier mag je SUBSELECTS bijvoorbeeld ook gebruiken in de FROM, SELECT of HAVING clausules.

#### Opdracht 112

Toon de titel en naam van alle boeken die bij hun uitgever de hoogste omzet hebben (verkocht * prijs)	

````python
%output 112
````

````python
%%uitwerking 112
SELECT boek.titel, uitgever.naam
FROM boek
JOIN uitgever
ON boek.uitgever = uitgever.code
WHERE (boek.verkocht * boek.prijs) = (
    SELECT MAX(b2.verkocht * b2.prijs)
    FROM boek b2
    WHERE b2.uitgever = boek.uitgever
)
ORDER BY uitgever.naam
````

#### Opdracht 113

Geef de titel van alle boeken die een coauteur hebben en niet het goedkoopste boek zijn	

````python
%output 113
````

````python
%%uitwerking 113
SELECT boek.titel
FROM boek
WHERE boek.isbn IN (SELECT boek FROM coauteur)
AND boek.prijs > (SELECT MIN(prijs) FROM boek)
ORDER BY boek.titel
````

#### Opdracht 114

geef van alle boeken de auteursnaam als ze een coauteur hebben met een naam die eerder in het alfabet voorkomt dan de naam van de auteur	

````python
%output 114
````

````python
%%uitwerking 114
SELECT DISTINCT a1.naam
FROM boek
JOIN auteur a1 ON boek.auteur = a1.code
JOIN coauteur ON boek.isbn = coauteur.boek
JOIN auteur a2 ON coauteur.auteur = a2.code
WHERE a2.naam < a1.naam
ORDER BY boek.titel, a1.naam;
````

## UNION

Met UNION kun je de resultaten van twee of meer queries tot 1 tabel samenvoegen. Een eis is wel dat beide queries hetzelfde aantal kolommen moeten hebben. Bij een UNION wordt altijd door SQL een DISTINCT uitgevoerd, dus dubbele rijen worden automatisch verwijderd. Als je de rijen in de uitvoer wilt sorteren, dan kun je dat alleen maar aan het eind van de gehele query doen over de eindtabel.

#### Opdracht 115

Geef (in een tabel) naam, adres, woonplaats van alle auteurs en uitgevers	

````python
%output 115
````

````python
%%uitwerking 115
SELECT naam, adres, woonplaats
FROM auteur
UNION
SELECT naam, adres, woonplaats
FROM uitgever
````

#### Opdracht 116

````python
%output 116
````

````python
%%uitwerking 116
SELECT naam, adres, woonplaats
FROM auteur
UNION
SELECT naam, adres, woonplaats
FROM uitgever
ORDER BY naam
````

#### Opdracht 117

Geef (in 1 tabel) naam, adres, woonplaats van alle auteurs die uit Den Haag komen en van alle uitgevers die een boek uitgeven van een auteur die uit Den Haag komt	

````python
%output 117
````

````python
%%uitwerking 117
SELECT naam, adres, woonplaats
FROM auteur
WHERE woonplaats == 'Den Haag'
UNION
SELECT uitgever.naam, uitgever.adres, uitgever.woonplaats
FROM uitgever
JOIN boek
ON boek.uitgever = uitgever.code
JOIN auteur
ON boek.auteur = auteur.code
WHERE auteur.woonplaats = 'Den Haag'
````

#### Opdracht 118

Geef de naam van de auteur die de meeste boeken heeft verkocht.

````python
%output 118
````

````python
%%uitwerking 118
SELECT auteur.naam
FROM auteur
JOIN boek
ON auteur.code = boek.auteur
GROUP BY auteur.naam
HAVING SUM(boek.verkocht) = 
    (SELECT MAX(totaal)
    FROM (
        SELECT SUM(verkocht) AS totaal
        FROM boek
        GROUP BY auteur))
````
