```

FROM artiest LEFT JOIN plaats ON woonplaats=plaats.id
```

Merk op dat MySQL je verplicht om het ON-gedeelte op te nemen. Bij standaard SQL is dat niet zo en kun je het ON-gedeelte weglaten, het resultaat is hetzelfde en daarmee is dit een veel gemaakte fout in SQL. 

````python
%%resultaat 224
Anouk, Den Haag
Direct, Den Haag
Kensington, Utrecht
Gordon, Amsterdam
````
