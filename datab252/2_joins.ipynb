## Unaire relaties

Een unaire relatie is een relatie van een tabel naar zichzelf. In dit voorbeeld is dat de koppeling tussen een leidinggevende (baas) en een ondergeschikte medewerker. Een unaire relatie werkt hetzelfde als een normale relatie, alleen moet je om 2 verschillende medewerkers te joinen de tabel medewerker hernoemen om onderscheid te maken tussen de baas en de ondergeschikte. Dat noemen we een alias.

#### Opdracht 72

Op <a href=#project>project</a>, geef per baas de naam en het aantal ondergeschikten, gesorteerd op naam van de baas	

````python
%output 72
````

````python
%%uitwerking 72
SELECT b.naam AS naam, COUNT(m.naam) AS 'aantal ondergeschikten'
FROM medewerker m
JOIN medewerker b
ON m.baas = b.nr
GROUP BY b.naam
````

````python
%%uitwerking 73
SELECT b.naam AS naam, COUNT(m.naam) AS "aantal ondergeschikten"
FROM medewerker m
JOIN medewerker b
ON m.baas = b.nr
WHERE m.woonplaats == 'Leiden'
GROUP BY b.naam
````

## Left Join

Een Inner Join tussen bijvoorbeeld medewerker en functie toont uiteindelijk alleen de medewerkers waar een functie bij hoort en functies waar een medewerker bij hoort. Soms wil je echter de resultaten van een bepaalde tabel behouden, ook als er geen rij aan gekoppeld kan worden uit de andere tabel. Daarvoor kun je gebruik maken van de Left Join. De Left Join behoud alle rijen uit de linker tabel (links van de left join operator) ook als die niet gejoined kunnen worden aan een rij uit de rechter tabel. We moeten daarvoor de koppeling tussen de twee tabellen (vaak Foreign Key = Primary Key) wel in de **ON** clausule opnemen. Voor rijen uit de linker tabel die niet gejoined kunnen worden aan de rechter tabel worden de kolommen uit de rechter tabel op NULL gezet. Merk op dat als je de where-clausule een hogere prioriteit heeft, dus rijen die met where worden uitgefilterd blijven weg.

#### Opdracht 89

Op <a href=#project>project</a>, toon van alle functies waar medewerkers bij horen, de functienaam en het totale inzet_perc van alle medewerkers met die functie	

````python
%output 89
````

````python
%%uitwerking 89
SELECT functie.naam, SUM(werkt.inzet_perc) AS 'totale inzet'
FROM functie
LEFT JOIN medewerker
ON medewerker.functie = functie.code
LEFT JOIN werkt
ON medewerker.nr = werkt.medewerker
WHERE medewerker.nr IS NOT NULL
GROUP BY functie.naam
````
