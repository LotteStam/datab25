#### Opdracht 324

#### Opdracht 325

Geef de naam van het team en de roepnaam van de aanvoerder van dat team, die een thuiswedstrijd hebben gespeeld op veldnr 2.

```python
%output 325
```

```python
%%uitwerking 325
SELECT DISTINCT
    team.teamnaam,
    lid.roepnaam
FROM team
JOIN wedstrijd
ON team.t_code = wedstrijd.t_code
JOIN thuis
ON wedstrijd.w_code = thuis.w_code
JOIN lid
ON team.aanvoerder = lid.l_code
WHERE thuis.veldnr = 2.0
```

#### Opdracht 327

```python
%%uitwerking 327
SELECT lid.roepnaam, wedstrijd.datum, team.teamnaam, dp_maker.aantal_dpn, wedstrijd.tegenstand
FROM dp_maker
JOIN lid
ON dp_maker.l_code = lid.l_code
JOIN wedstrijd
ON dp_maker.w_code = wedstrijd.w_code
JOIN team
ON wedstrijd.t_code = team.t_code
WHERE team.teamnaam LIKE 'Dames%'
```

#### Opdracht 328

Geef per team, de t_code , de teamnaam en het aantal wedstrijden tegen een tegenstander met het woord 'Klup' in de naam. Sorteer uitvoer oplopend op aantal wedstrijden.

```python
%output 328
```

```python
%%uitwerking 328
SELECT team.t_code, team.teamnaam, COUNT(wedstrijd.tegenstand)
FROM team
LEFT JOIN wedstrijd
ON team.t_code = wedstrijd.t_code AND wedstrijd.tegenstand LIKE '%Klup%'
GROUP BY team.t_code, team.teamnaam
```

### Opdracht 330

Geef per lid de roepnaam en achternaam en het totaal aantal doelpunten wat hij/zij gemaakt heeft wanneer dat er meer dan 1 is.

```python
%output 330
```

```python
%%uitwerking 330
SELECT lid.roepnaam, lid.achternaam, SUM(dp_maker.aantal_dpn)
FROM dp_maker
JOIN lid
ON dp_maker.l_code = lid.l_code
GROUP BY lid.roepnaam, lid.achternaam
HAVING SUM(dp_maker.aantal_dpn) > 1
```
